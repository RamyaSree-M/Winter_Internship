# Age column

```python
m=data['age'].mean()
```

```python
# Converting float into integer

data['age'] = data['age'].astype(int)
data['age']
```

# rating column

```python
# There is only 1 null value in rating column so, in order to handle it find mean

avg_rating = data['rating'].mean()
avg_rating
```

```python
# Filling the missing value with mean

data['rating'] = data.rating.fillna(avg_rating)
data['rating'] = data.rating.round(decimals = 1)

# Handle negative rating by replacing -1.0 with 1.0

data['rating'] = data['rating'].replace(-1.0,1.0)
data['rating']
```

# Salary column

```python
# Omitting '$' and convering in thousand

data['salary'] = data['salary'].str.replace('$','')
data['salary'] = data['salary'].str.replace('k','000')
data['salary']
```

```python
# Splitting salary by '-' into starting and ending salary

data[['starting_salary','ending_salary']] = data['salary'].str.split('-',expand = True)

# Deleting entire salary column
data = data.drop(columns= ['salary'])
```

# Easy Apply column

```python
# Replacing -1 with false

data['easy_apply'] = data['easy_apply'].replace('-1','FALSE')
data['easy_apply']
```

# Established Column

```python
# Finding mean to replace -1

avg_est = data['established'].mean()
avg_est
```

```python
data['established'] = data['established'].replace(-1,np.nan)
data['established'] = data['established'].replace(np.nan,avg_est)

# Converting data type into integer

data['established'] = data['established'].astype(int)
```

```python
data['established']
```

# Location column

```python
data['location'].unique()
```

```python
data['location'] = data['location'].str.replace(" ",",")
data['location'] = data['location'].str.replace("New,","New ")
```

```python
# Splitting location by ',' into country and country_ISO_code

data[['country','country_ISO_code']] = data['location'].str.split(",",expand = True)
data
```

```python
# After splitting the location column into 2 separted column, delete location column

data = data.drop(columns = ['location'])
```

```python
# Since there is default numbering so delete index column

data = data.drop(columns = ['index'])
```

# Calculate total revenue for each product

```python
# Calculating total revenue for each product
df['total_price'] = df['quantity'] * df['item_price']
revenue_per_product = df.groupby('item_name')['total_price'].sum()

print("Revenue from Each Product:")
print(revenue_per_product)
```

# Calculate total cost and quantity for each unique item

```python
# Calculating total cost and quantity for each unique item
df['total_cost'] = df['quantity'] * df['item_price']
item_summary = df.groupby('item_name').agg({
    'quantity': 'sum',
    'total_cost': 'sum'
}).reset_index()

print("Total Cost and Quantity of Every Unique Item:")
print(item_summary)
```

# Calculate avg quantity and avg cost per unit for each unique item

```python
# Calculating average quantity and average cost per unit for each unique item
item_summary = df.groupby('item_name').agg({
    'quantity': 'mean',
    'item_price': 'mean'
}).reset_index()

print("Average Quantity and Average Cost per Unit of Every Unique Item:")
print('\n',item_summary)
```
